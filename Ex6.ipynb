```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image1=cv2.imread('book.jpg')
image2=cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
kernel=np.ones((11,11),np.float32)/121
image3=cv2.filter2D(image2,-1,kernel)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image3)
plt.title('Filtered')
plt.axis('off')
```

```python
kernal2=np.array([[2,2,2],[4,8,4],[2,4,2]])/15
image3=cv2.filter2D(image2,-1,kernel)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image3)
plt.title('Weighted Averaging Filter')
plt.axis('off')
```

```python
gaussian_blur=cv2.GaussianBlur(src=image2,ksize=(11,11),sigmaX=0,sigmaY=0)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gaussian_blur)
plt.title('Gaussian Blur')
plt.axis('off')
```

```python
median=cv2.medianBlur(src=image2,ksize=7)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(median)
plt.title('Median Blurring')
plt.axis('off')
```

```python
kernel3=np.array([[1,2,1],[1,-5,1],[2,1,0]])
image3=cv2.filter2D(image2,-1,kernel3)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image3)
plt.title('Laplacian Kernal')
plt.axis('off')
```

```python
new_image=cv2.Laplacian(image2,cv2.CV_64F)

plt.figure(figsize=(9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title('Orignal')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(new_image)
plt.title('Laplacian Operator')
plt.axis('off')
```
